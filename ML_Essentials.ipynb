```python
harray=skills['Hardwork'].to_list()
sarray=skills['Skills'].to_list()
class km:
    def __init__(self,x,y):
        self.a1=x
        self.a2=y
    def randomcentroid(self):
        self.r=random.sample(list(range(len(self.a1)-1)),4)
        self.g=[[self.a1[self.r[0]],self.a2[self.r[0]]],[self.a1[self.r[1]],self.a2[self.r[1]]],[self.a1[self.r[2]],self.a2[self.r[2]]],[self.a1[self.r[3]],self.a2[self.r[3]]]]
    def cluster(self):
        self.arr=[[],[],[],[]]
        i=0
        while i<len(self.a1)-1:
            j=0
            while j<4:
                q=math.sqrt(((self.a1[i]-self.g[j][0])**2)+((self.a2[i]-self.g[j][1])**2))
                if j==0 or q>p:
                    p=q
                    l=j
                j=j+1    
            self.arr[l].append(i)
            i=i+1
            
    def newcentroid(self):
        m=0
        while m<4:
            n=0
            h=0
            w=0
            while n<len(self.arr[m]):
                h=h+self.a1[self.arr[m][n]]
                w=w+self.a1[self.arr[m][n]]
                n=n+1
            self.g[m][0]=h
            self.g[m][1]=w
            m=m+1
            
    def kmeans(self):
        self.randomcentroid()
        a=['A','B','C','D']
        v=0
        while v<1:
            self.cluster()
            self.newcentroid()
        g=0
        while g<4:
            b=0
            while b<len(self.arr[g])-1:
                lst1.append(self.a1[self.arr[g][b]])
                lst2.append(self.a2[self.arr[g][b]])
                b=b+1
            lst3.append(a[g]) 
            g=g+1
        data={"Hardwor":lst1,
              "skill":lst2,"grade":lst3 }
        df=pd.Dataframe(data)
                
pls=km(harray,sarray)
pls.kmeans()
        
            
        
        
    
        
        
    
```
